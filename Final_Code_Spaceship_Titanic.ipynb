## 2.2) Bivariate Analysis

```python
def analyze_passenger_data(feature_name, target_name = "Transported", target_value = True):
    """
    Analyzes and visualizes the relationship between a feature and passenger transport status.

    Parameters:
    - feature_name (str): The name of the feature to analyze.
    - target_name (str, optional): The name of the target column. Defaults to "Transported".
    - target_value (bool, optional): The value of the target to focus on. Defaults to True.
    """

    # Randomly select a colormap for the graph visualization to make each plot visually distinct
    colormaps = ['viridis', 'plasma', 'inferno', 'magma', 'cividis']
    rand_colormap = np.random.choice(colormaps)

    # Create a cross-tabulation between the feature of interest and the target variable,
    # then plot it as a bar chart for visual analysis
    transport_crosstab = pd.crosstab(df[feature_name], df[target_name])
    transport_crosstab.plot(kind = "bar", colormap = rand_colormap, figsize = (10, 4))
    plt.show()

    # Calculate the percentage of passengers transported (or not, based on target_value)
    # for each category of the feature. The result is sorted in descending order to identify
    # the categories with the highest rates of being transported.
    transported_rate = df.groupby(feature_name).apply(lambda x: 100 * np.mean(x[target_name] == target_value)).sort_values(ascending = False)

    # Print the transportation rate for the top 3 categories within the feature.
    # This helps in quickly identifying which categories have the highest influence on the target.
    for category, rate in transported_rate.head(3).items():
        print(f"Transportation rate for {category} in {feature_name}: {rate:.2f}%")
```

```python
def explore_continuous_feature(feature_name, target_name = "Transported", df = df):
    """
    Explores the relationship between a continuous feature and a binary target variable through
    statistical summaries, correlation analysis, and visualizations including density and box plots.

    Parameters:
    - feature_name (str): The name of the continuous feature to analyze.
    - target_name (str): The name of the target variable. Default is "Transported".
    - df (DataFrame): The pandas DataFrame containing the data. Must include both the feature and the target.

    The function prints summary statistics (mean, standard deviation, min, max, mode) for the specified feature,
    calculates and prints the correlation between the feature and the target, and analyzes the transportation
    rate for values above and below the feature's mean. It also generates a density plot and a box plot to
    visualize the distribution of the feature across different values of the target variable.
    """

    # Print summary statistics and mode for the specified feature
    print(f"Exploring {feature_name.upper()}...")
    print(df[feature_name].describe().drop(['count', '25%', '50%', '75%']))
    print(f"Mode of {feature_name}: {df[feature_name].mode()[0]}")

    # Calculate and print the correlation between the feature and the target variable
    corr = df[[feature_name, target_name]].corr().iloc[0, 1]
    print(f"Correlation with {target_name}: {corr:.3f}")

    # Determine and print transportation rates for values above and below the feature's mean
    above_mean = df[df[feature_name] > df[feature_name].mean()]
    below_mean = df[df[feature_name] < df[feature_name].mean()]
    above_ratio = above_mean[target_name].mean()
    below_ratio = below_mean[target_name].mean()
    print(f"Transportation Rate for Passengers Whose {feature_name} is ABOVE the Average Age of Passengers: {above_ratio:.3f}")
    print(f"Transportation Rate for Passengers Whose {feature_name} is BELOW the Average Age of Passengers: {below_ratio:.3f}")

    # Visualize the distribution and box plot of the feature, segmented by the target variable
    plt.figure(figsize=(14, 6))
    plt.subplot(1, 2, 1)
    sns.kdeplot(data=df, x=feature_name, hue=target_name, shade=True)
    plt.title("Density Plot")

    plt.subplot(1, 2, 2)
    sns.boxplot(data=df, y=feature_name, x=target_name)
    plt.title("Box Plot")
    plt.tight_layout()
    plt.show()
```

## 3.1) Basic Cleaning

```python
# Droping Irrelavant Columns
drop_col = df[["PassengerId", "Name", "Cabin"]]
df = df.drop(drop_col,axis=1)
```

## 3.2) Null Values Handling

```python
# Checking Null Values
df.isnull().sum()
```

```python
def impute_categorical(df, columns, target = 'Transported'):
    """
    Imputes missing values in categorical columns with the mode (most frequent value)
    within each category of the target variable.

    Parameters:
    - df (DataFrame): The pandas DataFrame containing the data.
    - columns (list): A list of column names (strings) to impute.
    - target (str): The name of the target variable based on which the mode is calculated.
                    Default is 'Transported'.

    Returns:
    - DataFrame: The DataFrame with imputed values for specified categorical columns.
    """
    for col in columns:
        # Calculate the mode for each category of the target variable
        mode_df = df.groupby(target)[col].apply(lambda x: x.mode()[0] if not x.mode().empty else None)

        # Impute missing values with the mode for each target category
        for target_value in df[target].unique():
            condition = (df[target] == target_value) & (df[col].isnull())
            df.loc[condition, col] = mode_df[target_value]
    return df
```

```python
def impute_numerical(df, columns, target='Transported'):
    """
    Imputes missing values in numerical columns with the mean value within each category
    of the target variable.

    Parameters:
    - df (DataFrame): The pandas DataFrame containing the data.
    - columns (list): A list of column names (strings) to impute.
    - target (str): The name of the target variable based on which the mean is calculated.
                    Default is 'Transported'.

    Returns:
    - DataFrame: The DataFrame with imputed values for specified numerical columns.
    """
    for col in columns:
        # Calculate the mean for each category of the target variable
        mean_df = df.groupby(target)[col].mean()

        # Impute missing values with the mean for each target category
        for target_value in df[target].unique():
            condition = (df[target] == target_value) & (df[col].isnull())
            df.loc[condition, col] = mean_df[target_value]
    return df
```

```python
# Columns in which Null values are present
categorical_columns = ['HomePlanet', 'Destination', 'CryoSleep', 'VIP']
numerical_columns = ['Age', 'RoomService', 'FoodCourt', 'ShoppingMall', 'Spa', 'VRDeck']

# Imputing Null Values
df = impute_categorical(df, categorical_columns)
df = impute_numerical(df, numerical_columns)
```

```python
df.isnull().sum()
```

## 3.3) Encoding

```python
# Encoding True False Features
true_false_mapper = {True: 1, False:0}

df.CryoSleep = df.CryoSleep.map(true_false_mapper)
df.VIP = df.VIP.map(true_false_mapper)
```

```python
# Encoding Nominal Features with OneHot Encoding
one_hot_encoder = OneHotEncoder()
one_hot_encoder.fit(df[["HomePlanet","Destination"]])

res = one_hot_encoder.transform(df[["HomePlanet","Destination"]]).toarray()
df[one_hot_encoder.get_feature_names_out()] = res
```

```python
# Droping Original columns after encoding
df.drop(["HomePlanet","Destination"], axis = 1, inplace = True)
```

## 3.4) Scaling

```python
# Applying MinMax Scaler
scaler = MinMaxScaler(feature_range=(0, 5))

# Storing columns in variable
num_cols = ["Age","RoomService","FoodCourt","ShoppingMall","Spa","VRDeck"]
scaler.fit(df[num_cols])
df[num_cols] = scaler.transform(df[num_cols])
```

# 4) Modeling

```python
# Assigning columns to x and y
x = df.drop("Transported",axis=1)
y = df['Transported']

# Splitting Data to Train and Test Part
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.2)
```

## 4.1) Experimenting with Multiple Models

```python
# Storing Created models in List
models = [
    SVC(),
    LogisticRegression(),
    KNeighborsClassifier(),
    DecisionTreeClassifier(),
    RandomForestClassifier(),
    HistGradientBoostingClassifier(),
    MultinomialNB()
]
```

```python
models_results = []

# Iterating, Training and Evaluting Models
for mdl in models:

    name = type(mdl).__name__

    # Training model
    mdl.fit(x_train, y_train)
    y_pred = mdl.predict(x_test)

    # Scores
    train_sco = mdl.score(x_train, y_train)
    test_sco = mdl.score(x_test, y_test)

    # Accuracy
    accuracy = accuracy_score(y_test, y_pred)

    # Precision
    precision = precision_score(y_test, y_pred, average = "macro")

    # Recall
    recall = recall_score(y_test, y_pred, average = "macro")

    # F1 Score
    f1_sco = f1_score(y_test, y_pred, average = "macro")

    # Classification Report
    report = classification_report(y_test, y_pred)

    # Confusion Matrix
    conf_mat = confusion_matrix(y_test, y_pred)

    # Storing Results
    res = []
    res.append(name)
    res.append(train_sco)
    res.append(test_sco)
    res.append(accuracy)
    res.append(precision)
    res.append(recall)
    res.append(f1_sco)
    models_results.append(res)

    # Printing Results
    print("+"*15, name, "+"*15)
    print("Training Score : ", train_sco)
    print("Testing  Score : ", test_sco)
    print('Accuracy : ', accuracy)
    print('F1 : ', f1_sco)
    print('Reacll : ', recall)
    print('Precision : ', precision)
    print('Classification Report....\n', report)
    plt.figure(figsize = (4, 3))
    sns.heatmap(conf_mat, annot = True, fmt = "d")
    plt.show()

    print("++++++++++++++++++++++++++++++++++++++++++++++++")
    print("++++++++++++++++++++++++++++++++++++++++++++++++")
```

```python
# Create a DataFrame from model evaluation and print the results
# This DataFrame includes the model names and their respective performance metrics:
# Training Score, Testing Score, Accuracy, Precision, Recall, and F1 Score.

model_results_df = pd.DataFrame(models_results, columns = ["Model", "Training Score", "Testing Score", "Accuracy", "Precision", "Recall", "F1 Score"])
model_results_df
```

Verification whether the models are overfitting or underfitting

```python
# Set the 'Model' column as the DataFrame's index and select only the 'Training Score' and 'Testing Score' columns for visualization.
# Plotting these scores using a bar chart, providing a visual comparison of training and testing performance across different models.

model_results_df.set_index("Model")[["Training Score", "Testing Score"]].plot(kind = "bar", figsize = (8, 4))
plt.show()
```

```python
# Checking Overfitness
model_results_df["Overfitness"] = model_results_df["Training Score"] - model_results_df["Testing Score"]
print(model_results_df["Overfitness"])

model_results_df.set_index("Model")[["Overfitness"]].plot(kind="bar", figsize=(8, 4))
plt.show()
```

HistGradient has the good accuracy and low overfitness so this is the best final model.

## 4.2) Final Model

```python
model = HistGradientBoostingClassifier()
model.fit(x_train, y_train)

y_pred = model.predict(x_test)

# Scores
train_sco = model.score(x_train, y_train)
test_sco = model.score(x_test, y_test)

# Accuracy
accuracy = accuracy_score(y_test, y_pred)

# Precision
precision = precision_score(y_test, y_pred, average="macro")

# Recall
recall = recall_score(y_test, y_pred, average="macro")

# F1 Score
f1_sco = f1_score(y_test, y_pred, average="macro")

# Classification Report
report = classification_report(y_test, y_pred)

# Confusion Matrix
conf_mat = confusion_matrix(y_test, y_pred)


# Printing Results
print("Training Score : ", train_sco)
print("Testing  Score : ", test_sco)
print('Accuracy : ', accuracy)
print('F1 : ', f1_sco)
print('Reacll : ', recall)
print('Precision : ', precision)
print('Classification Report....\n', report)
plt.figure(figsize=(4, 3))
sns.heatmap(conf_mat, annot=True, fmt="d")
plt.show()

print("++++++++++++++++++++++++++++++++++++++++++++++++")
print("++++++++++++++++++++++++++++++++++++++++++++++++")
```
